# Breast Cancer Dataset Exploration

In this notebook, we will load and explore the Breast Cancer dataset from scikit-learn.
The goal is to understand the structure of the dataset before using it for machine learning classification.

This dataset contains measurements computed from images of breast mass cell nuclei.

Each row represents one sample.

Each sample has 30 numerical features, such as mean radius, mean texture, mean perimeter, and mean area.

The target label has two classes:

- 0 = malignant
- 1 = benign

The goal is to use the 30 features to predict whether a tumor is malignant or benign.

A malignant tumor is cancerous, while a benign tumor is non-cancerous.


We use `load_breast_cancer()` from scikit-learn to load the dataset.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# Load the dataset
breast = load_breast_cancer()

# Create a DataFrame from the feature data
breast_dataset = pd.DataFrame(
    breast.data,
    columns=breast.feature_names
)

# Add the numeric label
breast_dataset["label"] = breast.target

# Add readable class names
breast_dataset["diagnosis"] = breast_dataset["label"].map({
    0: "Malignant",
    1: "Benign"
})

# Display first five rows
breast_dataset.head()
```

```python
print("Dataset shape:", breast_dataset.shape)
print("Number of samples:", breast_dataset.shape[0])
print("Number of features:", len(breast.feature_names))

print("\nTarget names:")
print(breast.target_names)
```

## Class Distribution

We count how many samples belong to each class.

This helps us understand whether the dataset is balanced or imbalanced.

```python
breast_dataset["diagnosis"].value_counts()
```
